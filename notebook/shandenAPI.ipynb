1. Make an API request that returns the months where "fire" was searched in 2016. Which month had the most searches?AUGUST 

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json'

params = {
    'query_text': 'fire',
    'year': '2016'
}

response = requests.get(endpoint, params = params)

response.text
```

```python
res = response.json()
```

```python
res
```

```python
pd.DataFrame(response.json())
```

2. Make an API request that returns all the times a query was run more than 100 times in a month. How many times did this occur?1000 

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?'
params = {"$where" : "query_count>=100"}

response2 = requests.get(endpoint, params = params)

response2.text
```

```python
response2 = pd.DataFrame(response2.json())
```

4. Make an API request that returns the entire Top 500 Monthly Searches dataset. Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.
